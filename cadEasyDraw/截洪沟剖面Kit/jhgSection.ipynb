```python
def sliceGroup(seq):
    a=[]
    count=len(seq)-1
    for i in range(count):
        j,*seq=seq
        a.append([j,seq[0]])
    return a
```

```python
def addColumn(acad,pnt,stylename,wordHeight=3.5):
    x,y,z=pnt
    acad.AddTable(Apoint(*pnt),7,1,10,35)
    text=['转点编号/桩号','平面距离（m）','自然地面标高',
         '沟顶标高(m)','设计沟内底标高（m）','坡度（%）','沟道型式及尺寸']
    for t in text:
        word=acad.AddText(t,Apoint(x,y-10*(text.index(t)+1)),wordHeight)
        word.StyleName=stylename
        
    
```

```python
addColumn(acad,pnt,'WLZ')
```
